```python
%matplotlib inline
from pandas.plotting import scatter_matrix
```

```python
scatter_matrix(df_iris[['sepal_length', 'sepal_width', 'petal_length']])
```

```python
df_iris.drop(['species'], axis=1).corr(method='spearman')
```

```python
from scipy import stats
stats.ttest_1samp(df_iris['sepal_length'], 3)
```

```python
#2-sample t-test: testing for difference across populations
setosaSL = df_iris[df_iris['species'] == 'setosa']['sepal_length']
versicolorSL = df_iris[df_iris['species'] == 'versicolor']['sepal_length']
#print(setosaSL)
#print(versicolorSL)
stats.ttest_ind(setosaSL, versicolorSL)
```

```python
df_iris[df_iris['species'] == 'setosa']['sepal_length']
```

```python
df_iris[df_iris['species'] == 'setosa']
```

```python
from statsmodels.formula.api import ols
```

```python
model = ols("petal_width ~ sepal_width", df_iris).fit()
```

```python
print(model.summary())
```

```python
# ANOVA
df_iris.boxplot("sepal_length", by = "species")
```

```python
import statsmodels.api as sm

mod = ols('sepal_length ~ species',
                data=df_iris).fit()
aov_table = sm.stats.anova_lm(mod, typ=2)
print (aov_table)
```

```python
# from sklearn. datasets import load_iris
from sklearn.decomposition import FactorAnalysis

X = df_iris.iloc[:, 0:4]
factor = FactorAnalysis(n_components=4, random_state=101).fit(X)
```

```python
print(pd.DataFrame(factor.components_, columns=X.columns))
```

```python
from sklearn.decomposition import PCA

pca = PCA().fit(X)
print( "Explained variance by component: %s" % pca.explained_variance_ratio_)
print("")
print (pd.DataFrame(pca.components_,columns=X.columns))
```
